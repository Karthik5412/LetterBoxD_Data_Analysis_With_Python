#### Droping Unwanted columns

```python
# Listing Columns to remove
un_cols = ['type','release_year','duration']
df = df.drop(columns= un_cols)
```

### Handling Null Values

```python
df.isna().sum()
```

```python
na_cols = ['director','cast','genres','description']
df = df.dropna(subset=na_cols)

df.isna().sum()
```

```python
# Filling Country column with mode value
df['country'] = df['country'].fillna(df['country'].mode()[0])
df.isna().sum()
```

```python
df['popularity'] = df['popularity'].replace(0,df['popularity'].mean())
df['vote_count'] = df['vote_count'].replace(0,df['vote_count'].mean())
df['budget'] = df['budget'].replace(0,df['budget'].mean())
df['vote_average'] = df['vote_average'].replace(0,df['vote_average'].mean())
df['revenue'] = df['revenue'].replace(0,df['revenue'].mean())
df['rating'] = df['rating'].replace(0,df['rating'].mean())
```

```python
df.head(2)
```

```python
def clean(x):
    data = x.split(',')
    in_data = []
    for i in data:
        i = i.strip()
        in_data.append(i)
    return (',').join(in_data)
```

```python
df['cast'] = df['cast'].apply(clean)
df['genres'] = df['genres'].apply(clean)
df['director'] = df['director'].apply(lambda x : x.strip())
```

```python
df['date_added'] = pd.to_datetime(df['date_added'])
df['year'] = df['date_added'].dt.year
df['month'] = df['date_added'].dt.month
df['day'] = df['date_added'].dt.day
df['date_added']= pd.to_datetime(df['date_added'])
df.head()
```

```python
countries_list = []
def clean(countries) :
    coun_list = countries.split(',')
    for country in coun_list :
        country = country.strip()
        if country not in countries_list:
            countries_list.append(country)
```

```python
df['country'].apply(clean)
len(countries_list)
```

```python
def actors(actors) :
    output = []
    actors_list = actors.split(',')
    for actor in actors_list :
        actor = actor.strip()
        output.append(actor)
    return output
        
```

```python
# df['cast'].apply(actors)
# len(unique_actors)
```

```python
df['cast'] = df['cast'].apply(actors)
df  = df.explode('cast').reset_index(drop=True)
```

```python
unique_gener = []    
def geners_list(gener) :
    gener_list = gener.split(',')
    for gener in gener_list :
        gener = gener.strip()
        if gener not in unique_gener:
            unique_gener.append(gener)
```

```python
df['genres'].apply(geners_list)
len(unique_gener)
```

```python
df.info()
```

```python
def collapse(x):
    data = x.split(',')
    result = []
    for i in data :
        result.append(i.strip())
    return result
        

df_geners = df.copy()
df_geners['genres'] = df_geners['genres'].apply(collapse)
df_geners_expanded = df_geners.explode('genres').reset_index(drop=True)

df_countries = df.copy()
df_countries['country'] = df_countries['country'].apply(collapse)
df_countries_expanded = df_countries.explode('country').reset_index(drop=True)
df_countries_expanded.head(2)
```

```python
df['genres'] = df['genres'].apply(collapse)
df = df.explode('genres').reset_index(drop=True)

df['country'] = df['country'].apply(collapse)
df = df.explode('country').reset_index(drop=True)

df['date_added'] = pd.to_datetime(df['date_added']).dt.strftime('%Y-%m')

df.head(2)

```

```python
df.to_csv('cleaned.csv')
```
